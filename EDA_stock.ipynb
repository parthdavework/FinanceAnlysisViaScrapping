```python
import matplotlib.dates as mdates
import plotly.graph_objects as go
from matplotlib.pyplot import figure


fig = go.Figure()
fig.update_layout(
            title='test',
            yaxis_title= 'TO Stock Price',
            shapes = [dict(
                x0='YYYY-MM-dd', x1='YYYY-MM-dd', y0=0, y1=1, xref='x', yref='paper',
                line_width=2)],
            annotations=[dict(
                x='YYYY-MM-dd', y=0.05, xref='x', yref='paper')]
)

figure(figsize=(12, 8), dpi=80)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
ax = plt.gca()
ax.tick_params(axis='x', labelrotation = 45)
plt.xlabel("Time with interval of 2 weeks")
plt.ylabel("Currency in dollars")
plt.title("Blackberry stock sale")

plt.plot(bb['date'], bb['high'], color='red')
plt.plot(bb['date'], bb['low'], color='green')
ax.legend(['high', 'low'])
```

```python


import matplotlib.dates as mdates
import plotly.graph_objects as go
from matplotlib.pyplot import figure

dict1={'Blackberry':bb,'Parkland Corporation':pki,'Denison Mines':dml,'Oncolytics Biotech Inc':onc,'Equitable Bank':eqb}

fig = go.Figure()
fig.update_layout(
            title='test',
            yaxis_title= 'TO Stock Price',
            shapes = [dict(
                x0='YYYY-MM-dd', x1='YYYY-MM-dd', y0=0, y1=1, xref='x', yref='paper',
                line_width=2)],
            annotations=[dict(
                x='YYYY-MM-dd', y=0.05, xref='x', yref='paper')]
)

for key,values in dict1.items():
    figure(figsize=(30, 8), dpi=80)
    #plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=360))
    ax = plt.gca()
    ax.tick_params(axis='x', labelrotation = 90)
    ax.ticklabel_format(style='plain')
    plt.xlabel("Time with interval of 2 weeks")
    plt.ylabel("Currency in dollars")
    plt.title("{} Date vs volume".format(key))
    sns.boxplot(values['date'], values['volume'])
    plt.show()
```

#Checking the date vs volume box plot and matching it with prices high and low

1. Blackberry: The stock sales for blackberry was highest on 30 Nov 2022 which was 6 times more than normal times and realised that the price gradually increased after some time after being low.

2. Dension Miles: The volume can be seen highest on 29 august 2022. By checking on the price we realised that the prices were peak during this time

3. Parkland corp: The volume was highest in 20 october 2022 and 30 nov. Again matching the high low in the graph we can clearly see that the prices have increased on both the dates after being low for some time.

4. Equitable Bank: Volumes high in between 20 and 22 nov. We had a price surge


# We noticed that the volume is directly proportional to the prices of the stock.When the prices increase the volumes go high as well in majority of the cases. People like to sell the stock when the prices are high to earn a profit, meanwhile some sell to decrease the margin of loss if paid higher price
